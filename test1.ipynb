```python
def word_frequency(text):
    words = text.split()
    freq = {}
    for word in words:
        word = word.lower()  
        if word in freq:
            freq[word] += 1
        else:
            freq[word] = 1
    return freq
print(word_frequency("CLOSE enough to win or win"))
```

```python
a = float(input("Введите первое число: "))
b = float(input("Введите второе число: "))
operation = input("Введите знак (+, -, *, /): ")

# Выполнение операции
if operation == "+":
    print("Результат:", a + b)
elif operation == "-":
    print("Результат:", a - b)
elif operation == "*":
    print("Результат:", a * b)
elif operation == "/":
    if b != 0:
        print("Результат:", a / b)
    else:
        print("Ошибка: деление на ноль!")
else:
    print("Неизвестная операция")
```

```python
def number_set(a, b):
    start = min(a, b)
    end = max(a, b)
    return set(range(start, end + 1))

result = number_set(7,4)
print(result)
```
